```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('cleanedData.csv')
```

```python
# Change categorical string to array
df['genres'] = df['genres'].apply(lambda x: x.split(","))
```

```python
y = []
x = {}
for item in df['genres']:
    y.append(len(item))
    for cat in item:
        x[cat] = '1'
        
# Maximum length of genres in one data
max(y)
```

```python
# All unique genres
x.keys()
```

```python
# All number of unique genres
len(x.keys())
```

```python
# Cast it to a dictionary
catDict = {element: idx+1 for idx, element in enumerate(x)}
print(catDict)
```

```python
import json 

# Serializing json
json_object = json.dumps(catDict, indent=4)
 
# Writing to genres.json
with open("genres.json", "w") as outfile:
    outfile.write(json_object)
```

```python
genres = df['genres'].apply(lambda x: [catDict[i] for i in x]).to_list()
```

```python
tempGenreArr = []
for row in genres:
    x = np.zeros(28)
    for genre in row:
        x[genre - 1] = 1
    tempGenreArr.append(x)
```

```python
transformedGenres = \
pd.DataFrame(tempGenreArr, columns=["genre"+element for idx, element in enumerate(catDict)])
```

```python
df = pd.concat([df.reset_index(), transformedGenres], axis=1)
```

```python
df.head()
```

```python
df = df.drop(['index'], axis=1)
```

```python
# Appply same dunction to directors column
# Change categorical string to array
df['directors'] = df['directors'].apply(lambda x: x.split(","))
```

```python
y = []
x = {}
for item in df['directors']:  
    y.append(len(item))
    for cat in item:
        x[cat] = '1'
```

```python
# Maximum length of directors in one row of data
max(y)
```

```python
tfArr = df['directors'].apply(lambda x: len(x) < 4)
```

```python
df = df.loc[tfArr]
```

```python
df.shape
```

```python
# Cast it to a dictionary
directorsDict = {element: idx+1 for idx, element in enumerate(x)}
directorsDict
```

```python
# Serializing json
json_object = json.dumps(directorsDict, indent=4)
 
# Writing to directors.json
with open("directors.json", "w") as outfile:
    outfile.write(json_object)
```

```python
len(directorsDict)
```

```python
directors = \
pd.DataFrame(df['directors'].apply(\
    lambda x: [\
    directorsDict[i] for i in x]).to_list(), columns=['dir1', 'dir2', 'dir3'])
```

```python
directors
```

```python
df = pd.concat([df.reset_index(), directors], axis=1)
```

```python
df.loc[:,['dir1', 'dir2', 'dir3']] = \
df.loc[:,['dir1', 'dir2', 'dir3']].fillna(0)
```

```python
df.drop(['genre1', 'genre2', 'genre3', 'dir1', 'dir2', 'dir3', 'index', 'tconst'], axis=1).to_csv('detailsDataset.csv', index=False)
```

```python
finaldf = df.drop(['genres', 'directors', 'index', 'tconst'], axis=1)
```
